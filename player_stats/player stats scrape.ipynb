```python
import requests
from bs4 import BeautifulSoup
```

```python
r = requests.get('http://www.espn.com/nba/statistics/player/_/stat/scoring-per-game/sort/avgPoints/seasontype/2/qualified/false')
```

```python
soup = BeautifulSoup(r.text,"html5lib")
```

```python
soup.find("table")
```

```python
table = soup.find("table")

headings = [th.get_text() for th in table.find("tr").find_all("td")]

datasets = []
for row in table.find_all("tr")[1:]:
    dataset = (td.get_text() for td in row.find_all("td"))
    datasets.append(dataset)

print(datasets)
```

```python
list(datasets[4])
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(datasets)
```

```python
df.columns = headings
```

```python
df.index = df.RK
```
