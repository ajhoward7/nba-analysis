# Scraping code experimentation

```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'http://www.espn.com/nba/news/archive'
```

```python
r = requests.get(url)
```

```python
soup = BeautifulSoup(r.text)
```

```python
links = soup.find_all("ul",{"class":"inline-list indent"})
```

```python
soup2 = BeautifulSoup(str(links[0]))
```

```python
for link in soup2.find_all("a"):
    print(link["title"],link["href"])
```

```python
articles = soup2.find_all("a")
```
